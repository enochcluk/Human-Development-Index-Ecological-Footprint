```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
footprints = pd.read_csv (r'C:/Users/enoch/Desktop/Earth/904.csv')
#sns.barplot(x=footprints.Region, y=footprints['Biocapacity Deficit or Reserve'])
#sns.scatterplot(x=footprints.Country, y=footprints['Biocapacity Deficit or Reserve'])
lisst=[]
for gdp in footprints['GDP per Capita']:
    zdp=str(gdp).replace(",","")
    gdp=str(zdp).replace("$","")
    lisst.append(float(gdp))
dat2 = pd.DataFrame({'GDP': lisst})
footprint=footprints.join(dat2)
footprint.drop(columns=['Countries Required'])

#'Cropland Footprint'
#Grazing Footprint
#Carbon Footprint
#GDP per Capita
#HDI
#Population (millions)
#Biocapacity Deficit or Reserve  
#Earths Required
```

```python
sns.scatterplot(x=footprints['GDP per Capita'], y=footprints['Carbon Footprint'], hue=footprints['Region'])
```

```python
#print(footprints)
corr=footprint.corr()
mask = np.zeros(corr.shape, dtype=bool)
mask[np.triu_indices(len(mask))] = True
plt.figure(figsize=(10,10))

thing = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap='rainbow',
    square=True,
    mask=mask
)
thing.set_xticklabels(labels=thing.get_xticklabels(),rotation=80)
#earths required- if everyone on earth lived like them, how many would we need?
```
